## Let's look at the responses to our survey using group and pivot

```python
link = 'https://docs.google.com/spreadsheets/d/10w89SKidIb2r2xzt_LW0JNeToPmGFg_FAM7PcbPWtbc/export?format=csv'

data_table = Table().read_table(link).drop('Timestamp')
data_table
```

### Group by the first variable to see how many people are in each category

```python
grouped_by_first = data_table.group(...)
grouped_by_first
```

### Group by the second variable to see how many people are in each category

```python
grouped_by_second = data_table.group(...)
grouped_by_second
```

### Let's see what happens when we put np.average into the group function

```python
grouped_by_second_avg = data_table.group(..., ...)
grouped_by_second_avg
```

### Group by the first variable and then the second

```python
grouped_by_first_second = data_table.group(...)
grouped_by_first_second
```

### How should we summarize our numerical variable besides a count?

```python
grouped_by_first_second_sum = data_table.group(make_array('phone_type', 'app_name'),max )
grouped_by_first_second_sum
```
